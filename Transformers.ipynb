```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

nltk.download('punkt')

# Example sentences
corpus = [
    "I love deep learning and natural language processing",
    "Word embeddings are fun to learn and visualize",
    "I enjoy solving problems using machine learning",
    "Chatbots are built using NLP techniques"
]

# Tokenize the sentences
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]
```

```python
# Train Word2Vec model
model_w2v = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=5, min_count=1)

# Get vector for a word
vector = model_w2v.wv['learning']
print("Vector for 'learning':\n", vector)

# Find similar words
print("\nWords similar to 'learning':")
print(model_w2v.wv.most_similar('learning'))
```

```python
similarity = model_w2v.wv.similarity('learning', 'machine')
print("\nSimilarity between 'learning' and 'machine':", similarity)
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer

dataset = load_dataset("imdb")
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
# Tokenization function
def tokenize_function(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=512)

# Apply tokenization
tokenized_datasets = dataset.map(tokenize_function, batched=True)

# Rename label column
tokenized_datasets = tokenized_datasets.rename_column("label", "labels")

# Set format for PyTorch
tokenized_datasets.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
import transformers
print(transformers.__version__)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    report_to="none",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_dir='./logs',
)

trainer = Trainer(

    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"].shuffle(seed=42).select(range(10000)),  # smaller for quick training
    eval_dataset=tokenized_datasets["test"].select(range(2000)),
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```
